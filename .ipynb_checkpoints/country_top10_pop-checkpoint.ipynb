```python
# dependecies
import pandas as pd
import numpy as np
```

```python
# File to load: world population from 1960 to 2018
# Note the file is also deposited on https://drive.google.com/drive/folders/1Y7rTb9--cxC-4FZwwuzI6-ermPAUVbPF?usp=sharing
population_load_data = (r"C:\Users\geofr\project1\world_population_2018aug.csv")

# Read the file and store into Pandas data frame
population_data = pd.read_csv(population_load_data)
population_data.head()
```

```python
# get the country name column and the year column
# year column can be changed based on the year required
# year available are from 1960 up to 2018

country_yr_df = population_data[["Country Name","2018"]]
country_yr_df = country_yr_df.sort_values(["2018"],ascending=False)

# type(country_yr_df)
country_yr_df.head()
```

```python
# get the top 10 most populated countries

country_df = country_yr_df[:10]
country_df 
```
